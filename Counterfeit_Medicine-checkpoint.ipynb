```python
y_train=ld_train["Counterfeit_Sales"]
x_train=ld_train.drop(["Counterfeit_Sales"],1)
y_test=ld_test["Counterfeit_Sales"]
x_test=ld_test.drop(["Counterfeit_Sales"],1)
```

```python
from sklearn.preprocessing import StandardScaler,Normalizer
```

```python
ss=StandardScaler()
```

```python
ss=Normalizer()
```

```python
ss.fit(x_train)
```

```python
x_train=ss.transform(x_train)
```

```python
x_test=ss.transform(x_test)
```

```python
lm=LinearRegression()
```

```python
lm.fit(x_train,y_train)
```

```python
lm.coef_
```

```python
p_test=lm.predict(x_test)

residual=p_test-y_test

rmse_lm=np.sqrt(np.dot(residual,residual)/len(p_test))

rmse_lm
```

```python
from sklearn.metrics import mean_absolute_error
MAE=mean_absolute_error(y_test,p_test)
Result=1-(MAE/1660)
```

```python
Result
```

```python
ridge=Ridge(fit_intercept=True,alpha=0.0001)

ridge.fit(x_train,y_train)

p_test=ridge.predict(x_test)

residual=p_test-y_test

rmse_ridge=np.sqrt(np.dot(residual,residual)/len(p_test))

rmse_ridge
```

# Lasso

```python
lasso=Lasso(fit_intercept=True,alpha=0.0001)

lasso.fit(x_train,y_train)

p_test=lasso.predict(x_test)

residual=p_test-y_test

rmse_lasso=np.sqrt(np.dot(residual,residual)/len(p_test))

rmse_lasso
```

```python
from sklearn.ensemble import RandomForestRegressor 
```

```python
rf.fit(x_train,y_train)
```

```python
p_test=rf.predict(x_test)

residual=p_test-y_test

rmse_rf=np.sqrt(np.dot(residual,residual)/len(p_test))

rmse_rf
```
